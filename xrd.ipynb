```python
from pymatgen.ext.matproj import MPRester
from pymatgen.entries.compatibility import MaterialsProjectCompatibility
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from IPython.display import Image
from pymatgen.analysis.diffraction.xrd import XRDCalculator
m = MPRester("OQSepLeUv7v95d4WW89s")
structure = m.get_structure_by_material_id("mp-24199")
struc =SpacegroupAnalyzer(structure).get_primitive_standard_structure()
xraycal = XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
xrdcalc = xraycal.get_pattern(structure , scaled=True, two_theta_range=(0, 90)) #motif !
xraycal.get_plot(structure, two_theta_range=(0, 90), annotate_peaks=True, ax=None, with_labels=True, fontsize=12)
miller_ind = xrdcalc.hkls
theta2 = xrdcalc.x
intensities = xrdcalc.y
```
